#### Develop a code using Matplotlib performing all Pyplot basics operation basic text and legend using Agriculture crop yield data set

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Datasets/yield_df.csv")

```

```python
#Basic Plot
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['hg/ha_yield'], marker='o', label='Crop Yield')
plt.title('Crop Yield Over Years')
plt.xlabel('Year')
plt.ylabel('Yield (hg/ha)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import warnings
# Suppress specific warning
warnings.filterwarnings("ignore", message="Creating legend with loc=\"best\" can be slow with large amounts of data.")
df = pd.read_csv("Datasets/yield_df.csv")
# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(df['Item'], df['hg/ha_yield'], color='skyblue', label='Yield')
plt.title('Crop Yield by Item')
plt.xlabel('Crop')
plt.ylabel('Yield (hg/ha)')
plt.xticks(rotation=45)
plt.legend()
plt.show()

```

```python
#Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(df['avg_temp'], df['hg/ha_yield'], color='red', label='Yield')
plt.title('Yield vs Average Temperature')
plt.xlabel('Average Temperature (C)')
plt.ylabel('Yield (hg/ha)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['hg/ha_yield'], bins=5, color='green', label='Yield')
plt.title('Distribution of Yields')
plt.xlabel('Yield (hg/ha)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
