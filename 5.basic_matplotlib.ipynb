#### Develop a code to perform Matplotlib functions to display all the basic plots

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
years = [1990, 1991, 1992, 1993, 1994]
values = [10, 20, 15, 25, 30]
categories = ['A', 'B', 'C', 'D', 'E']
category_values = [15, 35, 25, 30, 20]

# Line Plot
plt.figure(figsize=(10, 6))
plt.plot(years, values, marker='o')
plt.title('Line Plot')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(categories, category_values, color='skyblue')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(years, values, color='red')
plt.title('Scatter Plot')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(category_values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart')
plt.show()

# Box Plot
plt.figure(figsize=(10, 6))
data = [np.random.randn(100) for _ in range(5)]
plt.boxplot(data, labels=categories)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```
